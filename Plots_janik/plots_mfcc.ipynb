```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/Users/janikwahrheit/Library/CloudStorage/OneDrive-PersÃ¶nlich/01_Studium/01_Bachelor/05. WS_2024_25/Seminar RTN/Code/RTN-jakob/df_dataset_mfcc_merged.xlsx")
#drop columns
columns_to_drop = ['sample_pos', 'channel', 'date_time', 'daytime']
df.drop(columns=columns_to_drop, inplace=True)
#remove rows with no vehicle type 
df = df[df["vehicle"].notna()]
#change data type
df["vehilce"] = df["vehicle"].astype("category")
#remove outliers
df = df[df["mfcc_1"] >= 0]
```

```python
mfcc_columns = [col for col in df.columns if col.startswith("mfcc")]

fig, axes = plt.subplots(len(mfcc_columns), 1, figsize=(40,100))

for i, col in enumerate(mfcc_columns): 
    sns.boxplot(data=df, x="vehicle", y=df[col], ax=axes[i], palette=sns.color_palette("GnBu"))
    axes[i].set_title(f"Boxplot for {col}")
    axes[i].set_xlabel("vehicle type")
    axes[i].set_ylabel(col)
```

```python

fig, axes = plt.subplots(len(mfcc_columns), 1, figsize=(40,100))

for i, col in enumerate(mfcc_columns): 
    sns.boxplot(data=df, x="speed_kmh", y=df[col], ax=axes[i], palette=sns.color_palette("GnBu"))
    axes[i].set_title(f"Boxplot for {col}")
    axes[i].set_xlabel("speed_kmh")
    axes[i].set_ylabel(col)
```

```python

fig, axes = plt.subplots(len(mfcc_columns), 1, figsize=(40,100))

for i, col in enumerate(mfcc_columns): 
    sns.histplot(data=df, x=df[col], ax=axes[i], palette=sns.color_palette("GnBu"))
    axes[i].set_title(f"Histogram for {col}")
    axes[i].set_xlabel(col)
```
